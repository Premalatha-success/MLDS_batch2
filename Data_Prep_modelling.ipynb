### Data Preparation

```python
df.isnull().sum()
```

```python
age_median=df["Age"].median()
age_median
```

```python
df["Age"].replace(np.nan,age_median,inplace=True)
```

```python
df.isnull().sum()
```

```python
emb_mode=df["Embarked"].mode()[0]
emb_mode
```

```python
df["Embarked"].replace(np.nan,emb_mode,inplace=True)
```

```python
df.isnull().sum()
```

```python
duplicate=df.duplicated()
duplicate.sum()
```

```python
### Outlier 
sns.boxplot(x="Age",data=df)
```

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range=Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
lowAge,upAge = remove_outlier(df["Age"])
df["Age"]=np.where(df["Age"]>upAge,upAge,df["Age"])
df["Age"]=np.where(df["Age"]<lowAge,lowAge,df["Age"])
```

```python
sns.boxplot(x="Age",data=df)
```

### Dropping insignificant columns

### Modelling and Evaluation

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x=df.drop(["Survived"],axis=1)
y=df[["Survived"]]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
```

```python
model_lr=LogisticRegression()
```

```python
model_lr.fit(x_train,y_train)
```

```python
model_lr.score(x_train,y_train)
```

```python
model_lr.score(x_test,y_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,confusion_matrix
```

```python
predictions=model_lr.predict(x_test)
```

```python
conf_m = ConfusionMatrixDisplay(confusion_matrix(y_test,predictions))
conf_m.plot()
```
